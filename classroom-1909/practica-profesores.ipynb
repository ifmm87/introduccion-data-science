```python
import numpy as np
import pandas as pd

# uso de panda para la lectura de datos
encuesta = pd.read_csv('data/profesores_fin.csv')
normalista = encuesta['normalista'].values
normalista = np.count_nonzero(normalista == 1)

licenciatura = encuesta['licenciatura'].values
licenciatura = np.count_nonzero(licenciatura == 1)

especializacion = encuesta['especializacion'].values
especializacion = np.count_nonzero(especializacion == 1)

maestria = encuesta['maestria'].values
maestria = np.count_nonzero(maestria == 1)
labels = ['normalista', 'licenciatura', 'especializacion', 'maestria']
tipos = [ normalista, licenciatura, especializacion, maestria]
tipos
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig1, ax1 = plt.subplots()
explode = (0.1, 0, 0, 0)
ax1.pie(tipos, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
```

```python
import numpy as np
import pandas as pd

# uso de panda para la lectura de datos
encuesta = pd.read_csv('data/profesores_fin.csv')
computadora = encuesta['tiene_computadora'].values
si = np.count_nonzero(computadora == 1)
no = np.count_nonzero(computadora==0)
labels = ['Si ', 'No']
tipos = [ si, no]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig1, ax1 = plt.subplots()
explode = (0.1, 0)
ax1.pie(tipos, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
```
